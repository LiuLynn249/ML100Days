### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





````python
import numpy as np
import pandas as pd

L=np.arange(1,16).reshape(5,3)
i=['2013-01-01', '2013-01-02', '2013-01-03','2013-01-04','2013-01-05']
c=['A','B','C']
df=pd.DataFrame(L, index=i,columns=c)
print('原本df內容:')
print(df)
````

````python
df.loc[ '2013-01-01', 'A']
#取得X=A,Y=2013-01-01的值
````

````python
df.loc[ '2013-01-01', ['A', 'B'] ] 
#取得X=A~B,Y=2013-01-01的值
````

````python
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
#取得X=A,Y=2013-01-01~2013-01-02的值
````

````python
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
#取得X=A~C,Y=2013-01-01~2013-01-05的值
````

### 2. 請根據提供的資料，選擇出下列的要求：

```
# - select the first 3 rows.

# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件

import pandas as pd
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
# - select the first 3 rows.

df.head(3)
````

````python
# - select the odd rows. (index = 1, 3, 5)

df[df.index % 2 !=0]
````

````python
# - select the last 2 columns.
df[df.columns[-2:]]
````

````python
# - select the even columns. (index = 0, 2, 4)
df[df.columns[0::2]]
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
import pandas as pd
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
# - 1. filtered by first column > 20?
df[df.iloc[:,0]>20]
````

````python
# - 2. filtered by first column + second column > 50

df[df.iloc[:,0] + df.iloc[:,1]>50]
````

````python
# - 3. filtered by first column < 30 or second column > 30

df[(df.iloc[:,0]<30)|(df.iloc[:,1]>30)]
````

````python
# - 4. filtered by total sum of row > 100
df[df.sum(axis=1)>100]
````
