```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
map = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)

map.drawmapboundary()

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```

```python
map = Basemap(projection = 'mill', lat_0 = 0, lon_0 = 0,resolution ='l')

map.drawmapboundary()

# 畫圖
map.drawcoastlines()

map.drawparallels(range(-90,91,30), color='#CCCCCC')  # 緯線
map.drawmeridians(range(-180,181,60), color='#CCCCCC')  # 經線

# 顯示结果
plt.show()
```

```python
map = Basemap(projection = 'mill', lat_0 = 0, lon_0 = 0,resolution ='h')

map.drawmapboundary()

map.drawparallels(range(-90,91,30), color='#CCCCCC')  # 緯線
map.drawmeridians(range(-180,181,60), color='#CCCCCC')  # 經線

# 畫圖
map.drawcoastlines()

# 顯示结果
plt.show()
```
