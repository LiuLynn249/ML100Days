### 1. 請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。




```python
import numpy as np
Z = np.random.randint(0, 10, 1000000).reshape(100, 100, 100)
print(Z)
```

```python
%timeit -n 10 a = 2

for i in Z:
    for j in i:
        for k in j:
            j+=1
print(Z)
```

```python
%timeit -n 10 a = 2

for i in Z.flat:
    i+=1
print(Z)
```

```python
%timeit -n 10 a = 2

for i in np.nditer(Z):
    i=i+1
    
print(Z)
```

### 2. 如何從一個陣列中，找出出現頻率最高的數值與位置？


```python
import numpy as np
Z = np.random.randint(0,10,50)
print(Z)
```

```python
#Z.sort()
#print(Z)

key=np.unique(Z)    #取得有哪些數字
value=np.bincount(Z)#每個值的出現次數

dic=dict(zip(key.tolist(), value.tolist()))#變成dict
print(dic)

v=max(dic,key=dic.get)
print('頻率最高的數字:')
print(v)
print('頻率最高數字的位置:')
print(np.where(Z==v))
```

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


```python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))

b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))


c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
```

```python
def tolist(iterable):#by 遞迴+for迴圈
    #判斷是不是陣列
    if type(iterable) !=np.ndarray:
        return iterable
    
    newData=[]#給一個新的陣列
    for i in iterable:
        newData.append(tolist(i))
        #顯示程式如何轉 print(newData)
    return newData
    
print(tolist(a))
print(tolist(b))
print(tolist(c))


##與Day01的HW一樣....

```
