```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()

#畫出完整的 cos 圖形
'''
作業

'''
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

'''
#作業

#方法不只有一種

'''

plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
#請問在python中圖形Matplotlib如何顯示中文?    https://ithelp.ithome.com.tw/questions/10193102

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.title('散點圖: Scatter Plots')
plt.scatter(X,Y,color='b',alpha=0.5,s=5)
plt.show()
```

題目 : 將資料夾中boston.csv讀進來，並用圖表分析欄位。<br>
1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?<br>
2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?<br>

```python
import pandas as pd
import csv 

df=pd.read_csv("19_2_boston.csv") 
print(df)
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
column_name=list(df.columns)
df2=pd.DataFrame(df,columns=column_name)
df2.boxplot(fontsize=15,rot =30,figsize=(20,20))

c1=df2.median(axis=0)>300
c2=df2.median(axis=0)<400

c11=c1[c1==True].index
c22=c2[c2==True].index

print('中位數落在300~400的欄位=',c11.join(c22).values)
```

```python
#2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
X=df['NOX']
Y=df['DIS']
plt.scatter(X,Y,alpha=0.5,s=5,color='g')
print('負相關')
```
